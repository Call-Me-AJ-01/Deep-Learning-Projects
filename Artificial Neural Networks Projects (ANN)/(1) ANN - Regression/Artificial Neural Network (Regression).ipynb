### Package Used To Import Dataset

```python
import pandas as pd
```

### Importing the dataset

```python
data=pd.read_csv('50_Startups.csv')
data
```

### Checking For Null Values

```python
data.isnull().any()
```

### Counting The No Of Null Values On Each Columns

```python
data.isnull().sum()
```

### Viewing The Column Name

```python
data.columns
```

### Replacing The Null Values

```python
data['R&D Spend'].fillna(data['R&D Spend'].mean(),inplace=True)
data['Marketing Spend'].fillna(data['Marketing Spend'].mean(),inplace=True)
data['State'].fillna(data['State'].mode()[0],inplace=True)
data['Profit'].fillna(data['Profit'].mean(),inplace=True)
```

```python
data
```

### Checking For Null Values After Filling The Null Values

```python
data.isnull().any()
```

```python
data.isnull().sum()
```

### Seperating dependent and Independent Variable

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
x
```

```python
y
```

### Using OneHotEncoder and ColumnTransformer

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
col_trans=ColumnTransformer([('aj',OneHotEncoder(),[3])],remainder='passthrough')
x=col_trans.fit_transform(x)
```

```python
x
```

```python
y
```

### Splitting data's for training and testing Purpose

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

### Building The ANN Model

```python
from keras.models import Sequential
model=Sequential()
```

### Adding Input,Hidden,Output Layers

### Adding Input Layer

### Adding Output Layer

```python
# If activation argument is not given by default the Neural Network takes it as a regression Problem
model.add(Dense(units=1,init='random_uniform'))
```

### Compiling The Model

```python
model.compile(optimizer='adam',loss='mse',metrics=['mse'])
```

### Training The Model

```python
model.fit(x_train,y_train,epochs=2000,batch_size=8)
```

### Testing The Model By Predicting

```python
y_pred=model.predict(x_test)
y_pred
```
