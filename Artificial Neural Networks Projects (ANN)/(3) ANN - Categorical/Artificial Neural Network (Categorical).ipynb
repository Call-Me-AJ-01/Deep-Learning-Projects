### Converting The Output Column Into Unique Value Using 'OneHotEncoder'

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
z=data.iloc[:,:].values
```

```python
col_trans=ColumnTransformer([('aj',OneHotEncoder(),[0])],remainder='passthrough')
z=col_trans.fit_transform(z)
```

### Seperating Dependent and Independent Variable

```python
x=z[:,7:]
y=z[:,0:7]
```

### Splitting Data's For Training And Testing Purpose

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

### Building The Model
